```python
import pandas as pd # Основная библиотека для работы с таблицами
import numpy as np  # Библиотека для математических операций (нужна для обработки бесконечностей)
import re           # Регулярные выражения для очистки текста от лишних символов

# 1. ЗАГРУЗКА
f_data = pd.read_csv('financial_data.csv') #Функция read_csv открывает файл и переносит его в оперативную память.
p_data = pd.read_csv('prolongations.csv') #Мы создаем две переменные (f_data и p_data), чтобы работать с ними независимо.

# 2. ОЧИСТКА ДАННЫХ
# Превращаем ID в строки и убираем невидимые пробелы (.strip()).
# Это критично, чтобы ID '101' и '101 ' считались одним и тем же проектом при объединении.
f_data['id'] = f_data['id'].astype(str).str.strip() #Нормализация Ключей (ID) id — это мост, по которому мы соединим финансы и менеджеров.
p_data['id'] = p_data['id'].astype(str).str.strip() #o	.astype(str): Принудительно превращает всё в текст.
#.str.strip(): Удаляет невидимые пробелы по краям.

# Переводим названия всех колонок в нижний регистр (.lower()).
# Это защищает код от ошибок, если в одном файле написано "Январь", а в другом "январь".
f_data.columns = [str(c).strip().lower() for c in f_data.columns] #Это генератор списка (list comprehension).
#Он проходит по каждому названию колонки (for c in ...), убирает пробелы (strip) и делает буквы маленькими (lower).
p_data.columns = [str(c).strip().lower() for c in p_data.columns]

# Чистим данные внутри колонок 'month' и 'am'.
# .astype(str) гарантирует, что даже если в ФИО затесалось число, код не "упадет".
p_data['month'] = p_data['month'].astype(str).str.strip().str.lower()
p_data['am'] = p_data['am'].astype(str).str.strip()

# Создаем жесткий список месяцев. Это наш "каркас" для расчетов.
# Он нужен, чтобы код знал порядок месяцев (что после марта идет апрель).
all_months = [
    'ноябрь 2022', 'декабрь 2022', 'январь 2023', 'февраль 2023', 'март 2023',
    'апрель 2023', 'май 2023', 'июнь 2023', 'июль 2023', 'август 2023',
    'сентябрь 2023', 'октябрь 2023', 'ноябрь 2023', 'декабрь 2023', 'январь 2024', 'февраль 2024'
]#В цикле расчета мы будем говорить: «Возьми 5-й месяц из списка (Март), посчитай для него базу, а потом возьми 6-й месяц (Апрель)
#и посчитай для него базу, а потом возьми 6-й месяц (Апрель) и посчитай пролонгацию». Без этого списка расчет K1/K2 невозможен.

# ОБЪЕДИНЕНИЕ ТАБЛИЦ (JOIN)
# я приклеиваю финансовые данные к списку проектов пролонгации по ключу 'id'.
# left join гарантирует, что я не потеряю ни один проект из основного списка.
df = pd.merge(p_data[['id', 'month', 'am']],# pd.merge(...): Функция «склеивания» двух таблиц.
              f_data.drop(columns=['account', 'причина дубля'], errors='ignore'),
              on='id', how='left') #Мы выбрасываем из финансов столбцы с ФИО (так как в задании сказано, что данные из пролонгаций первичны)
#•	on='id', how='left': Соединяем по ключу id. Тип left означает:
#«Оставь все проекты из файла пролонгаций, и если по ним есть данные в финансах — приклей их рядом».
#создаем единую рабочую среду, где по каждому ID проекта сразу видно:
# кто за него отвечает, когда он закончился и сколько денег он приносил в каждый месяц.

# КОНВЕРТАЦИЯ СУММ В ЧИСЛА
for m in all_months:#Цикл перебирает каждый месяц из списка-каркаса
    if m in df.columns:
        # Заменяем запятую на точку и удаляем всё, кроме цифр и точек.
        # Это убирает значки рубля, пробелы-разделители и прочий "мусор".
        df[m] = pd.to_numeric(df[m].astype(str).str.replace(',', '.').apply(lambda x: re.sub(r'[^\d.]', '', x)), errors='coerce').fillna(0)
#Python понимает только точки в десятичных дробях.
results = []

# 3. РАСЧЕТ K1 И K2 (СЕРДЦЕ ЛОГИКИ)
for i in range(len(all_months)):
    curr_m = all_months[i] # Текущий месяц в цикле (T - месяц завершения)

    # К1 (Расчет за следующий месяц после завершения)
    if i + 1 < len(all_months):
        next_m = all_months[i+1] # Месяц T+1
        # Фильтруем только те проекты, которые завершились в curr_m
        base_1 = df[df['month'] == curr_m] #находим проекты, завершившиеся именно в этом месяце.
        if not base_1.empty:
            # .agg суммирует деньги и считает количество проектов (count) для базы.
            stats = base_1.groupby('am').agg({curr_m: 'sum', next_m: 'sum', 'id': 'count'})
            # Считаем "Успешно" — это количество строк, где отгрузка в следующем месяце > 0.
            success_count = base_1[base_1[next_m] > 0].groupby('am')['id'].count()
            for am in stats.index:
                if "2023" in next_m:#отсекаем лишние данные, записывая в итоговый список только показатели за 2023 год.
                    # Сама формула: Деньги(T+1) / Деньги(T)
                    val = (stats.loc[am, next_m] / stats.loc[am, curr_m]) if stats.loc[am, curr_m] > 0 else 0
                    results.append({'Менеджер': am, 'Период': next_m.capitalize(), 'Тип': 'K1', 'Коэффициент (%)': val, 'База (проектов)': stats.loc[am, 'id'], 'Успешно (шт)': success_count.get(am, 0)})
#K2 — это работа с теми, кто не продлился в первый месяц. Формула: Отгрузка T+2 / Отгрузка T.
    # К2 (Расчет за второй месяц после завершения)
    if i + 2 < len(all_months):
        next_m = all_months[i+1]
        after_next_m = all_months[i+2] # Месяц T+2
        # Условие по ТЗ: завершились в T, но нет отгрузки в T+1.
        base_2 = df[(df['month'] == curr_m) & (df[next_m] <= 0)]
#это самое важное. Мы берем проекты, завершившиеся в месяце T, у которых в месяце T+1 ноль отгрузки. Это те самые «отказники».
        if not base_2.empty:
            stats2 = base_2.groupby('am').agg({curr_m: 'sum', after_next_m: 'sum', 'id': 'count'})
            success_count2 = base_2[base_2[after_next_m] > 0].groupby('am')['id'].count()
            for am in stats2.index:
                if "2023" in after_next_m:
                    # Формула: Деньги(T+2) / Деньги(T)
                    val = (stats2.loc[am, after_next_m] / stats2.loc[am, curr_m]) if stats2.loc[am, curr_m] > 0 else 0
                    results.append({'Менеджер': am, 'Период': after_next_m.capitalize(), 'Тип': 'K2', 'Коэффициент (%)': val, 'База (проектов)': stats2.loc[am, 'id'], 'Успешно (шт)': success_count2.get(am, 0)})
#Все рассчитанные строки попадают в список results, который позже превратится в ваш финальный Excel-файл
# 4. СБОРКА ИТОГОВ
final_df = pd.DataFrame(results)
#Функция pd.DataFrame превращает этот список в полноценную таблицу.
# Расчет "ИТОГО ЗА ГОД" для каждого менеджера
yearly_am = final_df.groupby(['Менеджер', 'Тип']).agg({ #Применяем разные правила.
                                                       #Для коэффициента мы считаем среднее (mean), а количество проектов и успехов — суммируем (sum).
    'Коэффициент (%)': 'mean',
    'База (проектов)': 'sum',
    'Успешно (шт)': 'sum'
}).reset_index() #Группируем все строки одного человека (отдельно по K1 и K2).
yearly_am['Период'] = 'ИТОГО ЗА ГОД' #Создаем метку, чтобы в Excel эти строки было легко найти и отфильтровать.

# Расчет помесячных итогов для всего отдела
dept_monthly = final_df.groupby(['Период', 'Тип']).agg({
    'Коэффициент (%)': 'mean',
    'База (проектов)': 'sum',
    'Успешно (шт)': 'sum'
}).reset_index()
dept_monthly['Менеджер'] = 'ИТОГО ОТДЕЛ'

# Расчет годового итога для всего отдела (создаем ту самую недостающую переменную)
dept_yearly = dept_monthly.groupby('Тип').agg({
    'Коэффициент (%)': 'mean',
    'База (проектов)': 'sum',
    'Успешно (шт)': 'sum'
}).reset_index()
dept_yearly['Менеджер'] = 'ИТОГО ОТДЕЛ'
dept_yearly['Период'] = 'ИТОГО ЗА ГОД'

# Финальный штрих: теперь все переменные на месте
full_report = pd.concat([final_df, yearly_am, dept_monthly, dept_yearly], ignore_index=True)

# Сохранение в Excel
full_report.to_excel('Аналитический_отчет_2023_ИТОГ.xlsx', index=False)
print("Готово! Файл 'Аналитический_отчет_2023_ИТОГ.xlsx' создан успешно.")
```
